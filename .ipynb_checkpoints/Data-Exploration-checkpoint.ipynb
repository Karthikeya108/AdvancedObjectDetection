## Using object detection to extract information from image dataset

```python
def extract_info(images_path, file_path):
    labels = objDetector.yoloModelObj.LABELS
    with open(file_path, 'a') as f:
        for fn in os.listdir(images_path):
            if fn.endswith('png'):
                image = Image.open(os.path.join(images_path, fn))
                annotatedImage, boxes = objDetector.detect_obj(np.array(image.copy()))
                row = fn+','+str(len(boxes))+','
                for box in boxes:
                    row = row + labels[box.get_label()] + ','
                row = row[:-1]+'\n'
                f.write(row)
```

```python
extract_info(train_datapath, 'imageStats.csv')
```

```python
def csv_to_df(file_path):
    df = pd.DataFrame(columns=['img', 'num_objects', 'labels'])
    idx = 0
    with open(file_path) as f:
        for line in f:
            d = line.split(',')
            if len(d) < 3:
                d.append('na')
            df.loc[idx] = [d[0], int(d[1].replace('\n','')), '-'.join(d[2:len(d)]).replace('\n','')]
            idx += 1
    return df
```

```python
df = csv_to_df('imageStats.csv')
df.head(10)
```

## Analyzing the images using Computer Vision techniques

```python
image = cv2.imread('images/006918.png')
plt.figure(figsize=(20,12))
ax = plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
smooth = cv2.GaussianBlur(gray, (5,5), 0)
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 12))
f.tight_layout()

ax1.imshow(gray)
ax1.set_title('Gray scale image', fontsize=10)

ax2.imshow(smooth)
ax2.set_title('Gaussian smoothing', fontsize=10)
```

```python
canny = cv2.Canny(smooth, 10, 100)
plt.figure(figsize=(20,12))
ax = plt.imshow(canny)
```
