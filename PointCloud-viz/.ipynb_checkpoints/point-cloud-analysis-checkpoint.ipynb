```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os
from IPython.display import HTML
from point_cloud_utils import PointCloudVisualizer
```

### Load the lidar points from the bin file into a numpy array

```python
with open('0000000010.bin', 'rb') as f:
    lidar_points = np.fromfile(f, dtype=np.float32)
    print(lidar_points.shape)
```

### Reshape the array so that each row has `(x, y, z, r)` coordinates

```python
lidar_points = lidar_points.reshape((int(lidar_points.shape[0]/4), 4))
```

```python
lidar_points.shape
```

```python
train_datapath = '../data/velodyne/training/velodyne/'
vlp_streampath = '../data/2011_09_26/2011_09_26_drive_0005_sync/velodyne_points/data/'
```

### Initialize PointCloudVisualizer

```python
pcv = PointCloudVisualizer()
pcv.HORIZONTAL_RANGE = (-30, 30)
pcv.VERTICAL_RANGE = (-30, 30)
```

### Velodyne Lidar Point Cloud Bird's eye view

```python
HRES = 0.35                    # horizontal resolution (assuming 20Hz setting)
VRES = 0.4                     # vertical resolution
VFOV = (-24.9, 2.0)            # Field of view (-ve, +ve) along vertical axis
Y_FUDGE = 5                    # y fudge factor for velodyne HDL 64E
```

```python
fig = pcv.projection_bird_view_spectral(lidar_points, figsize=(8,8))
```

### Efficient implementation but without spectral mapping
#### Sliced channels (8) based on heights

```python
im = pcv.projection_birds_eye_multiple_channels(lidar_points, n_slices=6, height_range=(-2.0, 0.27), resolution=0.1)
fig, axes = plt.subplots(2, 3, figsize=(15,10), dpi=100)
axes = axes.flatten()
for i,ax in enumerate(axes):
    ax.imshow(im[:,:,i], cmap="gray", vmin=0, vmax=255)
    ax.set_facecolor((0, 0, 0))  
    ax.xaxis.set_visible(False)  
    ax.yaxis.set_visible(False)
    ax.set_title(i, fontdict={"size": 10, "color":"#FFFFFF"})
fig.subplots_adjust(wspace=0.20, hspace=0.20)
fig.show()
#fig.savefig('../images/bird_view_multichannel.png')
```

### Bird view single channel 

```python
#pcv.projection_bird_view(lidar_points, resolution=0.1, saveto='../images/birdview.png')
pcv.projection_bird_view(lidar_points, resolution=0.1, saveto=None)
```

#### Velodyne lidar point cloud front view projection with **height** as metric

```python
fig = pcv.projection_front_view(lidar_points, v_res=VRES, h_res=HRES, v_fov=VFOV, val="height", y_fudge=Y_FUDGE, figsize=(18,12))
#fig.savefig('../images/frontview_height.png')
```

#### Velodyne lidar point cloud front view projection with **depth** as metric

```python
fig = pcv.projection_front_view(lidar_points, v_res=VRES, h_res=HRES, v_fov=VFOV, val="depth", y_fudge=Y_FUDGE, figsize=(18,12))
#fig.savefig('../images/frontview_depth.png')
```

#### Velodyne lidar point cloud front view projection with **reflectence** as metric

```python
fig = pcv.projection_front_view(lidar_points, v_res=VRES, h_res=HRES, v_fov=VFOV, val="reflectance", y_fudge=Y_FUDGE, figsize=(18,12))
#fig.savefig('../images/frontview_reflectance.png')
```

```python
for fn in os.listdir(vlp_streampath):
    with open(os.path.join(vlp_streampath, fn), 'rb') as f:
        d = np.fromfile(f, dtype=np.float32)
        d = d.reshape((int(d.size/4), 4))
        im = pcv.projection_bird_view(d, resolution=0.1, saveto='./point-cloud-images/{}'.format(fn.replace('bin','png')))
```
